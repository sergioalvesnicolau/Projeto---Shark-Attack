```python
# Primeiro tratamento:

#> Garantir que o nome das colunas está limpo.
sadf.columns = sadf.columns.str.strip()

#> Novo dataframe sem as linhas nulas e com apenas algumas colunas selecionadas para facilitar a visualização.
sadf1 = sadf.loc[:6839, ['Date', 'Year', 'Type', 'Country', 'Area', 'Location', 'Activity',
                         'Name', 'Sex', 'Age', 'Injury', 'Fatal (Y/N)', 'Time', 'Species',
                         'Investigator or Source']]
```

```python
# Corrigir duplicação:
sadf1.drop(5217, inplace = True)

#> Ver se ainda resta alguma.
sadf1.duplicated().sum()
```

```python
# Explorar como está a primeira coluna a ser analisada, da fatalidade dos ataques:
#> Ver quantidade de valores nulos.
print(sadf1['Fatal (Y/N)'].isna().sum())

#> Ver quais os preenchimentos diferentes de Y e N.
sadf1['Fatal (Y/N)'].value_counts()
```

```python
#Tratamento do 'UNKNOWN':

#> Ver dados quais as lesões descritas para as fatalidades desconhecidas.
sadf1.loc[sadf1['Fatal (Y/N)'] == 'UNKNOWN']['Injury'].value_counts()
```

```python
#> Preencher a fatalidade das duas linhas que dizem 'No injury' com N.

#> Máscara em que a fatalidade é desconhecida e a lesão é 'No injury'.
mask1 = sadf1[(sadf1['Injury'] == 'No injury') & (sadf1['Fatal (Y/N)'] == 'UNKNOWN')]

#> Lista com os indíces das linhas dessa máscara.
noinjuryrows1 = list(mask1.index)

#> Preenchimento do 'Fatal (Y/N)' dessas linhas com N.
sadf1.loc[noinjuryrows1, 'Fatal (Y/N)'] = 'N'

#> Colocar as demais linhas com fatalidade desconhecida como valores nulos.
sadf1.loc[sadf1['Fatal (Y/N)'] == 'UNKNOWN', 'Fatal (Y/N)'] = np.nan
```

```python
# Limpando o resto da coluna inferindo a fatalidade:

# F:
sadf1[sadf1['Fatal (Y/N)'] == 'F'] #> Ver dados com 'F' em 'Fatal (Y/N)'
#> Provavelmente o F foi preenchido pensando em 'Fatal'. Os três são casos fatais.

# M:
sadf1[(sadf1['Fatal (Y/N)'] == 'M')] #> Ver dados com 'M' em 'Fatal (Y/N)'
sadf1.loc[92, 'Fatal (Y/N)'] = 'Y' #> Fatal
sadf1.loc[799, 'Fatal (Y/N)'] = 'N' #> Lesão indica que não foi fatal

# Nq:
sadf1[sadf1['Fatal (Y/N)'] == 'Nq'] #> Ver dado com 'Nq' em 'Fatal (Y/N)'
#> Lesão indica que não foi fatal.

# 2017:
sadf1[sadf1['Fatal (Y/N)'] == 2017] #> Ver dado com '2017' em 'Fatal (Y/N)'
sadf1.loc[sadf1['Fatal (Y/N)'] == 2017, 'Fatal (Y/N)'] = 'N' #> Lesão indica que não foi fatal.

# Y x 2:
sadf1[sadf1['Fatal (Y/N)'] == 'Y x 2'] #> Ver dado com 'Y x 2' em 'Fatal (Y/N)'
#> Injury: "Fatal"

# ' N', 'n', 'N ', 'y':
#> Tratamento de string para se juntarem a 'Y' e 'N'.

#Limpeza:
sadf1['Fatal (Y/N)'].replace({'F': 'Y', '.*[Nn].*': 'N', '.*[Yy].*': 'Y'}, regex = True, inplace = True)
```

```python
#Tratamento dos nulos na coluna:

#> Retirar as linhas em que a fatalidade e a lesão são valores nulos (sem possibilidade de inferência):
mask2 = (sadf1['Fatal (Y/N)'].isna() != 0) & (sadf1['Injury'].isna() != 0)
sadf1.drop(list(sadf1[mask2].index), inplace = True)

#> Checar quantas linhas ainda estão com fatalidade nula:
print(sadf1['Fatal (Y/N)'].isna().sum())

#> Ver quais as lesões para as fatalidades de valor nulo:

sadf1.loc[(sadf1['Fatal (Y/N)'].isna())]['Injury'].value_counts().head(10)
```

```python
#Tratamento para não fatais:
sadf1.loc[(sadf1['Fatal (Y/N)'].isna()) & 
          (sadf1['Injury'].str.contains('No injury|Survived|Non-fatal|Minor injuries', flags = re.IGNORECASE)),
          'Fatal (Y/N)'] = 'N'

#Tratamento para fatais:
sadf1.loc[(sadf1['Fatal (Y/N)'].isna()) & 
          (sadf1['Injury'].str.contains('Disappeared|Drowned|drowning|mortem|death|fatal|remains', flags = re.IGNORECASE)),
          'Fatal (Y/N)'] = 'Y'
```

```python
#Explorar a segunda coluna a ser analisada, do tipo de ataque.
print(sadf1['Type'].isna().sum())
sadf1['Type'].value_counts()
```

```python
# Limpando a coluna com os dados preenchidos.

# Watercraft:
#> Ver dados com 'Watercraft' em 'Type':
sadf1[sadf1['Type'] == 'Watercraft']
#> #> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Watercraft', 'Type'] = 'Invalid'

# Questionable:
#> Ver dados com 'Questionable' em 'Type'.
sadf1[sadf1['Type'] == 'Questionable']
#> #> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Questionable', 'Type'] = 'Invalid'

#Boat:
#> Ver dados com 'Boat' em 'Type'.
sadf1[sadf1['Type'] == 'Boat']
#> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Boat', 'Type'] = 'Invalid'

#?:
#> Ver dado com '?' em 'Type'.
sadf1[sadf1['Type'] == '?']
#> Preenchimento com 'invalid', já que não foi possível inferir tipo.
sadf1.loc[sadf1['Type'] == '?', 'Type'] = 'Invalid'

#Unconfirmed:
#> Ver dado com 'Unconfirmed' em 'Type'.
sadf1[sadf1['Type'] == 'Unconfirmed']
#> Preenchimento da linha que lesão indica ataque provocado:
sadf1.loc[255, 'Type'] = 'Provoked'
#>Preenchimento das demais com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Unconfirmed', 'Type'] = 'Invalid'

#Unverified:
#> Ver dado com 'Unverified' em 'Type'.
sadf1[sadf1['Type'] == 'Unverified']
#> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Unverified', 'Type'] = 'Invalid'

#Under investigation
#> Ver dado com 'Under investigation' em 'Type'.
sadf1[sadf1['Type'] == 'Under investigation']
#> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Under investigation', 'Type'] = 'Invalid'

#Sea Disaster
#> Ver dado com 'Sea Disaster' em 'Type'.
sadf1[sadf1['Type'] == 'Sea Disaster']
#> Preenchimento com 'Invalid', já que não foi possível inferir tipo ou se houve ataque de tubarão.
sadf1.loc[sadf1['Type'] == 'Sea Disaster', 'Type'] = 'Invalid'
```
